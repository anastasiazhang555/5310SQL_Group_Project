```python
##Q1: To calculate the average price per square foot for residential properties in each neighborhood.
#(1)Filter for residential use only, excluding mixed-use and commercial buildings.
#(2)Exclude invalid data where square footage is missing or zero.
#(3)Calculate:average sale_price / gross_square_feet per neighborhood.
#(4)Sort neighborhoods by average price per square foot, in descending order.

query1 = """
SELECT 
    n.neighborhood_name,
    COUNT(DISTINCT hl.property_id) AS num_properties,
    ROUND(AVG(s.sale_price / NULLIF(p.gross_square_feet, 0)), 2) AS avg_price_per_sqft
FROM house_locations hl
JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
JOIN sale s ON hl.property_id = s.property_id
JOIN properties p ON hl.property_id = p.property_id
WHERE 
    p.gross_square_feet IS NOT NULL AND p.gross_square_feet > 0
    AND (
        p.building_class_present ~ '^[ABC]'
        OR p.building_class_present IN ('R1', 'R2')
    )
    AND p.building_class_present NOT LIKE 'S%'
GROUP BY n.neighborhood_name
ORDER BY avg_price_per_sqft DESC;
"""

with engine.begin() as conn:
    df_neighborhood_prices = pd.read_sql(text(query1), conn)

df_neighborhood_prices
```

```python
##Q2: To calculate the average price per square foot for commercial properties in each neighborhood.
query2 = """
SELECT 
    n.neighborhood_name,
    COUNT(DISTINCT hl.property_id) AS num_properties,
    ROUND(AVG(s.sale_price / NULLIF(p.gross_square_feet, 0)), 2) AS avg_price_per_sqft
FROM house_locations hl
JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
JOIN sale s ON hl.property_id = s.property_id
JOIN properties p ON hl.property_id = p.property_id
WHERE 
    p.gross_square_feet IS NOT NULL AND p.gross_square_feet > 0
    AND p.building_class_present ~ '^[E-HK-O]'
    AND p.building_class_present NOT LIKE 'S%'
GROUP BY n.neighborhood_name
ORDER BY avg_price_per_sqft DESC;
"""

with engine.begin() as conn:
    df_commercial_prices = pd.read_sql(text(query2), conn)

df_commercial_prices
```

```python
# Q3：Average price per square foot by borough
query3 = """
SELECT
  n.borough_id,
  CASE
    WHEN n.borough_id = 1 THEN 'Manhattan'
    WHEN n.borough_id = 2 THEN 'Bronx'
    WHEN n.borough_id = 3 THEN 'Brooklyn'
    WHEN n.borough_id = 4 THEN 'Queens'
    WHEN n.borough_id = 5 THEN 'Staten Island'
    ELSE 'Other'
  END AS borough_name,
  COUNT(DISTINCT hl.property_id) AS num_properties,
  ROUND(AVG(s.sale_price / NULLIF(p.gross_square_feet, 0)), 2) AS avg_price_per_sqft
FROM house_locations hl
JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
JOIN sale s ON hl.property_id = s.property_id
JOIN properties p ON hl.property_id = p.property_id
WHERE 
  p.gross_square_feet IS NOT NULL 
  AND p.gross_square_feet > 0
  AND (
    p.building_class_present ~ '^[ABC]' 
    OR p.building_class_present IN ('R1', 'R2')
  )
  AND p.building_class_present NOT LIKE 'S%'
GROUP BY n.borough_id
ORDER BY avg_price_per_sqft DESC;
"""
with engine.begin() as conn:
    df_borough_prices = pd.read_sql(text(query3), conn)
df_borough_prices
```

```python
# Q4: Calculate the average price per square foot by borough (for commercial use)

query3 = """
SELECT
  n.borough_id,
  CASE
    WHEN n.borough_id = 1 THEN 'Manhattan'
    WHEN n.borough_id = 2 THEN 'Bronx'
    WHEN n.borough_id = 3 THEN 'Brooklyn'
    WHEN n.borough_id = 4 THEN 'Queens'
    WHEN n.borough_id = 5 THEN 'Staten Island'
    ELSE 'Other'
  END AS borough_name,
  COUNT(DISTINCT hl.property_id) AS num_properties,
  ROUND(AVG(s.sale_price / NULLIF(p.gross_square_feet, 0)), 2) AS avg_price_per_sqft
FROM house_locations hl
JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
JOIN sale s ON hl.property_id = s.property_id
JOIN properties p ON hl.property_id = p.property_id
WHERE 
  p.gross_square_feet IS NOT NULL 
  AND p.gross_square_feet > 0
  AND p.building_class_present ~ '^[E-HK-O]'  -- ⚡改成商业用途筛选
  AND p.building_class_present NOT LIKE 'S%'
GROUP BY n.borough_id
ORDER BY avg_price_per_sqft DESC;
"""

with engine.begin() as conn:
    df_borough_prices_commercial = pd.read_sql(text(query3), conn)

df_borough_prices_commercial
```

```python
#Q5:
#Find out which properties meet both of the following criteria: (quality education resources+ convenient transportation)
#input the zipcode
#Within 500 meters of a subway station 
#Within 500 meters of a school 
#The property has a transaction record (from the sale table)

zipcode = '11106'

query5 = f"""
WITH property_points AS (
    SELECT 
        hl.property_id,
        ul.latitude AS prop_lat,
        ul.longitude AS prop_lon,
        s.sale_price
    FROM house_locations hl
    JOIN unified_locations ul ON hl.location_id = ul.location_id
    JOIN sale s ON s.property_id = hl.property_id
    WHERE ul.zipcode = '{zipcode}'
),

station_points AS (
    SELECT 
        sl.station_id,
        ul.latitude AS station_lat,
        ul.longitude AS station_lon
    FROM station_location sl
    JOIN unified_locations ul ON sl.location_id = ul.location_id
    WHERE ul.zipcode = '{zipcode}'
),

school_points AS (
    SELECT 
        sl.ats_system_code,
        ul.latitude AS school_lat,
        ul.longitude AS school_lon
    FROM school_locations sl
    JOIN unified_locations ul ON sl.location_id = ul.location_id
    WHERE ul.zipcode = '{zipcode}'
),

property_near_station AS (
    SELECT DISTINCT p.property_id
    FROM property_points p
    JOIN station_points s ON
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS((s.station_lat - p.prop_lat)/2)), 2) +
            COS(RADIANS(p.prop_lat)) * COS(RADIANS(s.station_lat)) *
            POWER(SIN(RADIANS((s.station_lon - p.prop_lon)/2)), 2)
        )) < 500
),

property_near_school AS (
    SELECT DISTINCT p.property_id
    FROM property_points p
    JOIN school_points s ON
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS((s.school_lat - p.prop_lat)/2)), 2) +
            COS(RADIANS(p.prop_lat)) * COS(RADIANS(s.school_lat)) *
            POWER(SIN(RADIANS((s.school_lon - p.prop_lon)/2)), 2)
        )) < 500
)

SELECT 
    p.property_id,
    p.sale_price,
    p.prop_lat,
    p.prop_lon,
    pr.year_built,
    pr.residential_units,
    pr.land_square_feet,
    pr.gross_square_feet
FROM property_points p
JOIN property_near_station ps ON p.property_id = ps.property_id
JOIN property_near_school sc ON p.property_id = sc.property_id
JOIN properties pr ON p.property_id = pr.property_id
ORDER BY p.sale_price DESC
LIMIT 20;
"""

with engine.begin() as conn:
    df_nearby_homes_full = pd.read_sql(text(query5), conn)

df_nearby_homes_full
```

```python
# Set the CSV file save path
csv_filename = 'Users/celine/Desktop/residential_with_qualify_resources.csv'

# SQL query: Added SELECT DISTINCT to avoid duplicate property_id entries
query_all_zipcodes = """
WITH property_points AS (
    SELECT 
        hl.property_id,
        ul.latitude AS prop_lat,
        ul.longitude AS prop_lon,
        ul.zipcode,
        s.sale_price
    FROM house_locations hl
    JOIN unified_locations ul ON hl.location_id = ul.location_id
    JOIN sale s ON s.property_id = hl.property_id
),

station_points AS (
    SELECT 
        ul.latitude AS station_lat,
        ul.longitude AS station_lon
    FROM station_location sl
    JOIN unified_locations ul ON sl.location_id = ul.location_id
),

school_points AS (
    SELECT 
        ul.latitude AS school_lat,
        ul.longitude AS school_lon
    FROM school_locations sl
    JOIN unified_locations ul ON sl.location_id = ul.location_id
),

property_near_station AS (
    SELECT DISTINCT p.property_id
    FROM property_points p
    JOIN station_points s ON
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS((s.station_lat - p.prop_lat)/2)), 2) +
            COS(RADIANS(p.prop_lat)) * COS(RADIANS(s.station_lat)) *
            POWER(SIN(RADIANS((s.station_lon - p.prop_lon)/2)), 2)
        )) < 500
),

property_near_school AS (
    SELECT DISTINCT p.property_id
    FROM property_points p
    JOIN school_points s ON
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS((s.school_lat - p.prop_lat)/2)), 2) +
            COS(RADIANS(p.prop_lat)) * COS(RADIANS(s.school_lat)) *
            POWER(SIN(RADIANS((s.school_lon - p.prop_lon)/2)), 2)
        )) < 500
)

SELECT DISTINCT
    p.property_id,
    ROUND(p.sale_price / NULLIF(pr.gross_square_feet, 0), 2) AS price_per_sqft,
    p.prop_lat AS latitude,
    p.prop_lon AS longitude,
    p.zipcode
FROM property_points p
JOIN property_near_station ps ON p.property_id = ps.property_id
JOIN property_near_school sc ON p.property_id = sc.property_id
JOIN properties pr ON p.property_id = pr.property_id
WHERE
    (
      pr.building_class_present ~ '^[ABC]'
      OR pr.building_class_present IN ('R1', 'R2')
    )
    AND pr.building_class_present NOT LIKE 'S%'
    AND pr.gross_square_feet > 0
ORDER BY price_per_sqft;
"""

# Execute the query and load the result into a DataFrame
with engine.begin() as conn:
    df_all_zip = pd.read_sql(text(query_all_zipcodes), conn)
```

```python
#Q7: Compare property prices near schools vs. not near schools in each borough.
#1.Get all sold properties with location and borough info.
#2.Get all schools' locations.
#3.Check distance: mark properties within 500m of any school as “Near School.”
#4.Group by borough and proximity to schools.
#5.Calculate average price and count for each group.
#6.Map borough IDs to names in Python.
query7 = """
WITH property_points AS (
    SELECT 
        hl.property_id,
        ul.latitude AS prop_lat,
        ul.longitude AS prop_lon,
        s.sale_price,
        p.gross_square_feet,
        n.borough_id
    FROM house_locations hl
    JOIN unified_locations ul ON hl.location_id = ul.location_id
    JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
    JOIN sale s ON hl.property_id = s.property_id
    JOIN properties p ON hl.property_id = p.property_id
    WHERE 
        p.gross_square_feet IS NOT NULL AND p.gross_square_feet > 0 AND
        (
            p.building_class_present ~ '^[ABC]' OR
            p.building_class_present IN ('R1', 'R2')
        )
        AND p.building_class_present NOT LIKE 'S%'
),

school_points AS (
    SELECT 
        ul.latitude AS school_lat,
        ul.longitude AS school_lon
    FROM school_locations sl
    JOIN unified_locations ul ON sl.location_id = ul.location_id
),

nearby_properties AS (
    SELECT DISTINCT p.property_id
    FROM property_points p
    JOIN school_points s
      ON ABS(s.school_lat - p.prop_lat) < 0.02
     AND ABS(s.school_lon - p.prop_lon) < 0.02
    WHERE 
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS((s.school_lat - p.prop_lat)/2)), 2) +
            COS(RADIANS(p.prop_lat)) * COS(RADIANS(s.school_lat)) *
            POWER(SIN(RADIANS((s.school_lon - p.prop_lon)/2)), 2)
        )) < 500
)

SELECT 
    p.borough_id,
    CASE WHEN p.property_id IN (SELECT property_id FROM nearby_properties)
         THEN 'Near School'
         ELSE 'Not Near School' END AS school_proximity,
    ROUND(AVG(p.sale_price / NULLIF(p.gross_square_feet, 0)), 2) AS avg_price_per_sqft,
    COUNT(*) AS num_properties
FROM property_points p
GROUP BY p.borough_id, school_proximity
ORDER BY p.borough_id, school_proximity;
"""

with engine.begin() as conn:
    df_school_borough = pd.read_sql(text(query7), conn)

# borough 映射
borough_map = {
    1: "Manhattan",
    2: "Bronx",
    3: "Brooklyn",
    4: "Queens",
    5: "Staten Island"
}
df_school_borough["borough_name"] = df_school_borough["borough_id"].map(borough_map)
df_school_borough[["borough_name", "school_proximity", "avg_price_per_sqft", "num_properties"]]
```

```python
#Q10:Recommend residential properties based on user preferences.
#Filters only housing units (codes starting with A, B, C, R1, R2; excludes mixed-use like S).
#Lets users set:
#(1)price range
#(2)minimum number of residential units
#(3)location (borough)
#(4)whether it's near a school or subway
#Uses Haversine formula to check if a property is within 500 meters of a school or subway.

Returns top N matching properties sorted by sale price.
def recommend_properties(
    min_price=None,
    max_price=None,
    min_units=None,
    near_school=False,
    near_station=False,
    borough_id=None,
    limit=10
):
    filters = []

    # Core housing use filter
    housing_condition = """
        (
            p.building_class_present ~ '^[ABC]' OR
            p.building_class_present IN ('R1', 'R2')
        )
        AND p.building_class_present NOT LIKE 'S%'
    """
    filters.append(housing_condition)

    # Other conditions
    if min_price is not None:
        filters.append(f"s.sale_price >= {min_price}")
    if max_price is not None:
        filters.append(f"s.sale_price <= {max_price}")
    if min_units is not None:
        filters.append(f"p.residential_units >= {min_units}")
    if borough_id is not None:
        filters.append(f"n.borough_id = {borough_id}")

    base_condition = " AND ".join(filters)

    # Haversine computation snippet
    haversine = """
        6371000 * 2 * ASIN(SQRT(
            POWER(SIN(RADIANS(({lat2} - ul.latitude)/2)), 2) +
            COS(RADIANS(ul.latitude)) * COS(RADIANS({lat2})) *
            POWER(SIN(RADIANS(({lon2} - ul.longitude)/2)), 2)
        )) < 500
    """

    # Is it close to school?
    if near_school:
        base_condition += f""" AND EXISTS (
            SELECT 1
            FROM school_locations sl2
            JOIN unified_locations ul2 ON sl2.location_id = ul2.location_id
            WHERE {haversine.format(lat2='ul2.latitude', lon2='ul2.longitude')}
        )"""

    # Is it close to a subway station?
    if near_station:
        base_condition += f""" AND EXISTS (
            SELECT 1
            FROM station_location stl
            JOIN unified_locations ul3 ON stl.location_id = ul3.location_id
            WHERE {haversine.format(lat2='ul3.latitude', lon2='ul3.longitude')}
        )"""

    # Final SQL queries
    query = f"""
    SELECT 
        hl.property_id,
        s.sale_price,
        p.year_built,
        p.residential_units,
        p.land_square_feet,
        p.gross_square_feet,
        p.building_class_present,
        ul.latitude,
        ul.longitude,
        n.borough_id
    FROM house_locations hl
    JOIN unified_locations ul ON hl.location_id = ul.location_id
    JOIN neighborhoods n ON hl.neighborhood_id = n.neighborhood_id
    JOIN sale s ON hl.property_id = s.property_id
    JOIN properties p ON hl.property_id = p.property_id
    WHERE {base_condition}
    ORDER BY s.sale_price DESC
    LIMIT {limit}
    """

    with engine.begin() as conn:
        result = pd.read_sql(text(query), conn)

    return result
```
